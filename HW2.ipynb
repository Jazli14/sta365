### Make the "Classic Bayes" figure, but instead of a single likelihood, plot the individual likelihoods of each observation
I will be using 3 likelihoods

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(10000)

# Generate a small random dataset
sample_size = 3
data = stats.norm(loc=0, scale=1).rvs(sample_size)

# Define parameters for the prior distribution
prior_center, prior_precision = 4, 1/2

# Specify the range of parameter values
parameter_values = np.linspace(-10, 10, 1001)

# Initialize an array for individual likelihoods
individual_likelihoods = np.zeros((sample_size, len(parameter_values)))

# Calculate individual likelihoods for each observation
for i in range(sample_size):
    individual_likelihoods[i] = stats.norm.pdf(data[i], loc=parameter_values, scale=1)

# Calculate the prior distribution
prior_distribution = stats.norm(loc=prior_center, scale=1/np.sqrt(prior_precision)).pdf(parameter_values)

# Compute the joint likelihood by multiplying individual likelihoods
joint_likelihood = np.prod(individual_likelihoods, axis=0) * prior_distribution * 500

# Calculate the posterior distribution
posterior_mean = (prior_precision * prior_center + np.sum(data)) / (prior_precision + sample_size)
posterior_scale = 1 / np.sqrt(prior_precision + sample_size)
posterior_distribution = stats.norm(loc=posterior_mean, scale=posterior_scale).pdf(parameter_values)

# Start Plotting
plt.figure(figsize=(16, 5))
for i in range(sample_size):
    plt.plot(parameter_values, individual_likelihoods[i], label="Observation " + str(i + 1) + " likelihood")
plt.plot(parameter_values, prior_distribution, label="Prior")
plt.plot(parameter_values, joint_likelihood, label="Joint (times scalar 500)")
plt.plot(parameter_values, posterior_distribution, label="Posterior")
plt.legend()
plt.title("Classic Bayes with Individual Likelihood")
plt.show()
```

### Make the "Classic Bayes" figure for the normal-normal model on a log scale

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(10000)

# Generate random data
n = 3
x = stats.norm(loc=0, scale=1).rvs(n)

# Define prior parameters
theta_0, tau = 4, 1/2

# Parameter domain
linspace_domain = np.linspace(-10, 10, 1001)

# Calculate log likelihood
likelihood = np.sum(stats.norm.logpdf((x[:, np.newaxis]) * np.ones(linspace_domain.shape),
                                      loc=linspace_domain, scale=1), axis=0)

# Calculate log prior
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).logpdf(linspace_domain)

# Calculate log posterior
posterior_mean = (tau * theta_0 + x.sum()) / (tau + n)
posterior_scale = 1 / (tau + n)**0.5
posterior = stats.norm(loc=posterior_mean, scale=posterior_scale).logpdf(linspace_domain)

# Calculate log joint
joint = likelihood + prior

# Plot results
plt.figure(figsize=(16, 5))
plt.plot(linspace_domain, likelihood, label="Log Likelihood")
plt.plot(linspace_domain, prior, label="Log Prior")
plt.plot(linspace_domain, posterior, label="Log Posterior")
plt.plot(linspace_domain, joint, linestyle="--", label="Log Joint")
plt.legend()
plt.title("Classic Bayes on a Log Scale")
plt.show()
```

### What is the posterior distribution of a normal likelihood with $p(\theta) \propto 1$?
Since we know $p(\theta | x) = \frac{p(\theta)p(x|\theta)}{p(x)}$ which means $p(\theta | x) \propto p(\theta)p(x|\theta)$

Therefore similar to the proof above we will expand the square term.

$$=\text{m}\times\bigg[\prod_{i=1}^n\sqrt{\frac{\phi}{2\pi}} e^{-\frac{\phi\left(x_i-\theta\right)^2}{2}}\bigg]$$

$$\text{Where m is a constant.}$$

$$\propto\exp\bigg[-\frac{1}{2}(\sum_{i=1}^n \phi(x_i-\theta)^2)\bigg]$$

$$\propto \text{exp}\bigg(-\frac{1}{2}\left[\phi\sum^n_{i=1}(x_i^2-2x_i\theta+\theta^2)\right]\bigg)$$

$$\propto \text{exp}\bigg(-\frac{1}{2}\left[\phi\sum^n_{i=1}x_i^2-2\theta\phi\sum^n_{i=1}x_i+n\phi\theta^2\right]\bigg)$$

$$\text{Since the first term does not have $\theta$ we ignore it.}$$

$$\propto\text{exp}\bigg(-\frac{1}{2}\left[n\phi\theta^2-2\theta\phi\sum^n_{i=1}x_i\right]\bigg)$$

$$\text{Then we can add back in a $\frac{\phi}{n}\sum^n_{i=1}x_i^2$ term to make a perfect square since its proptional}$$

$$\propto\text{exp}\bigg(-\frac{1}{2}\left[n\phi\theta^2-2\theta\phi\sum^n_{i=1}x_i + \frac{\phi}{n}\sum^n_{i=1}x_i^2\right]\bigg)$$

$$\propto\text{exp}\bigg(-\frac{1}{2}n\phi\left[\theta^2-\frac{2\theta}{n}\sum^n_{i=1}x_i+\frac{1}{n^2}(\sum^n_{i=1}x_i)^2\right]\bigg)$$

$$\propto \text{exp}\bigg(-\frac{1}{2}n\phi\left[\theta-\frac{1}{n}\sum^n_{i=1}x_i\right]^2\bigg)$$


Therefore we can see that the result is a normal distribution


### How do log likelihoods and posteriors differ from each other as a result of improper priors and marginal likelihood constants?

The posterior distribution is equal to the likelihood $\times$ prior $\div$ marginal likelihood
$$\Rightarrow p(\theta | x) = \frac{p(x | \theta)p(\theta)}{p(x)}$$

Therefore taking the log of the posterior looks like this
$$ \Rightarrow \log\big(p(\theta | x)\big)$$
$$ = \log \bigg(\frac{p(x | \theta)p(\theta)}{p(x)}\bigg)$$

Using log rules we can break it down.

$$ = \log\big(p(x | \theta)\big) + \log\big(p(\theta)\big) - \log\big(p(x)\big) $$

By the definition of an improper prior \log(p(\theta)) is considered small or irrelevant and may lack a finite value therefore we can remove the term.

$$ \Rightarrow \log\big(p(\theta | x)\big) = \log\big(p(x | \theta)\big) - \log\big(p(x)\big) $$

Demonstrating the the log posterior differs from the log likelihood through the $\log(\text{marginal likelihood})$ or $\log\big(p(x)\big)$.
